```python
from cvgeomkit.utils.helpers import order_clockwise
from cvgeomkit.geometry.points import Point
from shapely.geometry import Polygon
```

```python
poly = Polygon([(1, 3), (3, 3), (1, 1), (3, 1)])
poly.is_valid
```

```python
poly
```

```python
p1 = Point(1, 3)
p2 = Point(3, 3)
p3 = Point(3, 1)
p4 = Point(1, 1)
```

```python
area = [
    p1,
    p2,
    p4,
    p3,
]
```

```python
order_clockwise(area)
```

```python
poly = Polygon(order_clockwise(area))
poly.is_valid
```
